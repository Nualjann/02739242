```python
import networkx as nx
import matplotlib.pyplot as plt

def bfs(graph, start_node):
    visited = set()
    queue = [(start_node, None)]  # (node, parent)
    bfs_tree = nx.Graph()

    while queue:
        current_node, parent = queue.pop(0)
        if current_node not in visited:
            visited.add(current_node)
            if parent is not None:
                bfs_tree.add_edge(parent, current_node)
            neighbors = graph.neighbors(current_node)
            for neighbor in neighbors:
                if neighbor not in visited:
                    queue.append((neighbor, current_node))

    return bfs_tree

# Example usage:
# Create a graph with a cycle
nodes = ['F', 'I', 'J', 'K', 'L', 'M']
edges = [('F', 'I') , ('F', 'K'),  ('I', 'L'),('I', 'K'), ('J', 'L'), ('J','M') , ('L', 'M')]
G = nx.Graph()
G.add_nodes_from(nodes )
G.add_edges_from(edges)

# Perform BFS starting from node 'A' and generate the BFS tree
bfs_tree = bfs(G, 'F')


# Visualize the original graph, BFS tree, and DFS tree
plt.figure(figsize=(15, 5))

# Visualize the original graph
plt.subplot(131)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=12, font_weight='bold')
plt.title("Original Graph")

# Visualize the BFS tree
plt.subplot(132)
pos_tree = nx.spring_layout(bfs_tree)
nx.draw(bfs_tree, pos_tree, with_labels=True, node_color='lightgreen', node_size=500, font_size=12, font_weight='bold')
plt.title("BFS Tree")



plt.tight_layout()
plt.show()
```
